```python
# k plus proches voisins avec sklearn
# séparer les données
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# Définir le classifieur
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
# Définir le jeu d'apprentissage
clf.fit(X_train, y_train)
```

```python
# Afficher les résultat
print("Prediction : \n{}".format(clf.predict(X_test)))
```

```python
# Afficher la précision des résultats
print("Score : {}".format(clf.score(X_test, y_test)))
```

```python
# Afficher le résultat attendu
print("Resultat attendu : \n{}".format(y_test))
```

```python
# Visualiser la frontière de décision
import matplotlib.pyplot as plt
ig, axes = plt.subplots(1, 4, figsize=(12,3))
# remplir les plots
for n, ax in zip([1,3,9,15], axes):
    clf = KNeighborsClassifier(n_neighbors=n).fit(X, y)
    # séparateur
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:,0], X[:,1], y, ax=ax)
    ax.set_title("{} neighbor(s)".format(n))
    ax.set_xlabel("Caractéristique 0")
    ax.set_ylabel("Caractéristique 1")
axes[0].legend(loc=3)
```

```python
# Comparaison de la précision de l'apprentissage en fonction des k voisins
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
# Charger et diviser la BDD
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer['data'], cancer['target'], stratify=cancer.target, random_state=66)
# Déclarations des listes et de la portée de k

training_accuracy = []
test_accuracy = []
neighbors_settings = range(1,11)

for nb in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=nb).fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))

print(training_accuracy)
    
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("k voisins")
plt.legend()
                            
```

```python
# Avec sklearn
from sklearn.neighbors import KNeighborsRegressor

X, y = mglearn.datasets.make_wave(n_samples=40)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state =0)

reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)
```

```python
print("Prediction :\n {}".format(reg.predict(X_test)))
print("Attendu :\n{}".format(y_test))
print("Score total : {}".format(reg.score(X_test, y_test)))
```

```python
# visualiser l'équilibre
fig, axes = plt.subplots(1, 3, figsize=(15,4))
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
for n, ax in zip([1,3,9], axes):
    reg = KNeighborsRegressor(n_neighbors=n)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)
    ax.set_xlabel("target")
    ax.set_ylabel("features")
    ax.set_title("{} neighbor(s)\nTrain score : {}\nTest Score : {}".format(
        n, reg.score(X_train, y_train), reg.score(X_test, y_test)))
axes[0].legend(["Model prediction", "Train data", "Test data"])
```
